```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.get('https://en.wikipedia.org/wiki/Nathuram_Godse')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup
```

```python
heading = soup.find('h1').text
```

```python
heading
```

```python
print(soup.text.replace('\n\n', ''))
```

```python
soup.find_all('p')
```

```python
for p in soup.find_all('p'):
    print(p.text)
    print('-' * 10)
```

```python
corpus = ' '

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'

corpus = corpus.strip()
```

```python
print(corpus)
```

```python
fd = open(heading + '.txt', 'w')
fd.write(corpus)
fd.close()
```
